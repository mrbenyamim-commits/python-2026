```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

juegos = ["cuphead", "ender ring", "megaman 2", "minecraft", "fornike"]
numeros = [56, 44, 31, 12, 81]

plt.figure(figsize=(6, 3))
plt.plot(juegos, numeros, color="gold", marker="o", linestyle="--", linewidth=2.8)

plt.title("cantidad de compras por dia", fontsize=15 ,fontweight='bold')
plt.xlabel("juegos", fontsize=15)
plt.ylabel("compras por dia", fontsize=15)
plt.grid(True, linestyle="--", alpha=2.8)

plt.show()
```
