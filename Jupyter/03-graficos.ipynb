### **Gráficos Básicos en Matplotlib**

#### **Importando las librerías de Matplotlib y numpy**

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### **1. Gráfico de Líneas (plt.plot)**

**¿Para qué sirve?**

Se utiliza principalmente para mostrar la evolución o tendencia de una o más variables continuas a lo largo del tiempo o de una secuencia ordenada.

Para crear este tipo de gráfico, pasamos una secuencia de valores para el eje $X$ y otra para el eje $Y$. Matplotlib unirá de forma secuencial cada par de puntos $(x, y)$ mediante un segmento de línea recta.

```python
# Creación de dos listas en python con meses y valores de venta
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
ventas = [15, 24, 18, 32, 28, 40]

# Creación de la figura y el gráfico
plt.figure(figsize=(8, 4)) 
plt.plot(meses, ventas, color='green', marker='o', linestyle='--', linewidth=2.5)

# Personalización Gráfico
plt.title('Evolución Mensual de Ventas - 1° Semestre', fontsize=14)
plt.xlabel('Meses', fontsize=10)
plt.ylabel('Cantidad de Ventas (en miles)', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.8) 

# Mostrar el gráfico
plt.show()
```

### **2. Gráfico de Barras (plt.bar)**

**¿Para qué sirve?**  

Es ideal para comparar magnitudes o cantidades entre diferentes categorías discretas. Permite identificar rápidamente cuál categoría es la mayor, la menor o cómo se distribuyen los valores.

Para crear un gráfico de barra se requiere un conjunto de categorías (en el eje $X$) y sus respectivos valores numéricos (altura de las barras en el eje $Y$).

```python
lenguajes = ['Python', 'C++', 'Java', 'JavaScript', 'C#']
preferencia = [45, 20, 15, 12, 8] # Porcentajes ficticios solo con motivos de crear el gráfico

plt.figure(figsize=(8, 4))
plt.bar(lenguajes, preferencia, color=['indigo', 'orange', 'red', 'gold', 'teal'])

plt.title('Preferencia de Lenguajes de Programación 2026', fontsize=15, fontweight='bold')
plt.xlabel('Lenguajes de Programación', fontsize=15)
plt.ylabel('Porcentaje de Preferencia (%)', fontsize=15)

plt.show()
```

### **3. Gráfico de Dispersión (plt.scatter)**

**¿Para qué sirve?**  

Se utiliza para analizar la relación o correlación entre dos variables numéricas continuas. Permite observar si visualmente existe un patrón, una tendencia directa, inversa o si los datos se comportan de forma aleatoria (nube de puntos).

Para construir un gráfico de dispersión, cada dato se dibuja como un punto independiente en sus coordenadas de plano cartesiano $(x, y)$, sin unirse mediante líneas.

```python
# Se genera datos ficticios usando NumPy
# Se simula las horas de estudio (X) versus Calificación obtenida (Y) de 30 estudiantes
np.random.seed(42) 
""" Con seed() se establece una semilla para generar n° aleatorios
La semilla es un valor que controla el patrón de aleatoriedad, de modo que 
si se ejecuta el mismo código varias veces con la misma semilla, se obtendrán los mismos números aleatorios """

horas_estudio = np.random.uniform(2, 12, 30)
# El método uniform genera números aleatorios que siguen una distribución uniforme continua. 
# Esto significa que todos los números dentro del rango especificado tienen exactamente la misma probabilidad de ser elegidos
calificacion = 1.5 * horas_estudio + np.random.normal(0, 1.5, 30)
calificacion = np.clip(calificacion, 1.0, 7.0) # Ajusta a la escala de notas chilena
# A diferencia de los dos anteriores, clip() no genera datos nuevos, sino que revisa un arreglo existente 
# y "recorta" (o limita) sus valores para que se mantengan estrictamente dentro de un rango obligatorio.

plt.figure(figsize=(8, 5))
plt.scatter(horas_estudio, calificacion, color='green', edgecolors='darkgreen', s=80, alpha=0.8)

plt.title('Relación entre Horas de Estudio y Calificación', fontsize=14)
plt.xlabel('Horas Semanales de Estudio Autónomo', fontsize=12)
plt.ylabel('Nota Final (1.0 - 7.0)', fontsize=12)
plt.axhline(4.0, color='red', linestyle=':', label='Nota de Aprobación (4.0)') # Línea de aprobación
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```

### **4. Histograma (plt.hist)**

**¿Para qué sirve?**

Sirve para representar la distribución de frecuencias de una sola variable cuantitativa continua. Nos muestra cómo se agrupan los datos, dónde está el centro de los mismos y qué tanta dispersión o variabilidad existe (por ejemplo, si siguen una campana de Gauss).

Para crear este gráfico se divide el rango de los datos en "intervalos" o contenedores (bins) y cuenta cuántos elementos caen dentro de cada intervalo, dibujando una barra proporcional a ese conteo.

```python
# Se genera datos usando NumPy (Distribución normal de tiempos de ejecución de un algoritmo)
tiempos_ejecucion = np.random.normal(loc=150, scale=25, size=200) # 200 ejecuciones en milisegundos
# Este método genera números aleatorios que siguen una distribución normal (también conocida como distribución gaussiana)
# A diferencia de la uniforme, aquí los valores no tienen la misma probabilidad: los números cercanos al centro (la media) 
# son muy probables, y se vuelven cada vez más raros a medida que se alejan.
plt.figure(figsize=(8, 4))
plt.hist(tiempos_ejecucion, bins=15, color='seagreen', edgecolor='white', alpha=0.95)

plt.title('Distribución de Tiempos de Ejecución del Algoritmo', fontsize=14)
plt.xlabel('Tiempo de Ejecución (milisegundos)', fontsize=12)
plt.ylabel('Frecuencia (Cantidad de veces)', fontsize=12)
plt.grid(True, axis='y', alpha=0.25)

plt.show()
```
