## contexto del problema: 

la direccion general del aguas (DGA) de la region de gamboa
$m^3/s$

## requisitos:

escribir un codigo utilizando 2 funciones definidas que cumpla las siguientes espesificaciones:

#### 1. funcion: `procesar_cauda()`
- no debe recibir parametros de entrada
- utiliza una lista para almacenar caudales registrados
- utiliza un arreglo unidimecional de numby para generar la secuencia de horas de 1 al 8
- construir y devolver un dataframe de oandas que contenga dos columnas estruradas por hora y cauda

#### 2. funcion: `visualizar_curva(dataframe)`
- debe recibir como parametro el dataframe generado por la funcion anterior.
- utilizando *matplotlib*, generar un grafico de lineas que debe obserbar la curva de crecida del rio
- el grafico debe incorporar:
    - una linea con marcador en forma diamante en cada medicion `(marker="o")`
    - un titulo descriptibo `"monitoreo de caudal rio gamboa"`
    - etiqueta en los eje x


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def procesar_cauda():
    horas = np.arange(1,9)
    caudales = [12.5, 14.2, 18.1, 22.8, 25.4, 21.0, 17.5, 15.3]
    data = {
        "Hora": horas,
        "Caudal": caudales
    }
    # volver diccionario a dataframe:
    df = pd.DataFrame(data)
    return df
```

```python
def visualizar_curva(dataframe):
    fig, ax = plt.subplots()

    ax.plot(dataframe["Hora"], dataframe["Caudal"], marker='p', color='red', linestyle="--")
    ax.set_title("Monitoreo de caudal rio gamboa", fontsize=20)
    ax.set_xlabel("tiempo (horas)", fontsize=15)
    ax.set_ylabel("caudal (m^3/s)", fontsize=15)

    ax.grid(True)
```

```python
dat = procesar_cauda()
visualizar_curva(dat)
plt.show()
```
