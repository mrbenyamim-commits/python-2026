```python
import numpy as np
import matplotlib.pyplot as plt

lenguajes = ['Python', 'C++', 'Java', 'JavaScript', 'C#']
preferencia = [45, 20, 15, 12, 8] # Porcentajes ficticios solo con motivos de crear el gráfico

print(f"{lenguajes} \n{preferencia}")
```

```python
plt.figure(figsize=(4,3))
plt.bar(lenguajes, preferencia, color=['teal', 'blue', 'orange', 'gold', 'indigo'])

plt.title('Preferencia de Lenguajes de Programación 2026', fontsize=10, fontweight='bold')
plt.xlabel('Lenguajes de Programación', fontsize=10)
plt.ylabel('Porcentaje de Preferencia (%)', fontsize=10)

plt.show()
```
